```python
#Single line comment
letter = 'P'    # A string could be a single character or a bunch of texts
print(letter)    #P
print(len(letter))  #1
greeting = 'Hello, World!'  #String could be a single or diuble quote, "Hello, World!"
print(greeting)
print(len(greeting))  #13
sentence = 'I hope you are enjoying python classes'
print(sentence)
print(len(sentence))
```

```python
#Multiline string
multiline_string = '''I am a teacher and enjoy teaching.
I didn't find anything as rewarding as empowering people.
That is why I created python basic.'''
print(multiline_string)
#Another way of doing same thing
multilinestring = """I am a teacher and enjoy teaching.
I didn't find anything as rewarding as empowering people.
That is why I created python basic."""
print(multilinestring)
```

```python
#String concatenation
first_name= 'Pratibha'
last_name= 'Gunwannt'
space= ' '
fullname= first_name + space + last_name
print(fullname)   #Pratibha Gunwannt
#Checking length of a string using len() builtin function.
print(len(first_name))   #8
print(len(last_name))
print(len(first_name) == len(last_name))       #equal
print(len(first_name) > len(last_name))
print(len(fullname))
```

```python
#unpacking characters
language = 'Python'
a, b, c, d, e, f = language   #unpacking sequence characters into variables
print(a)    #P
print(b)    #y
print(c)
print(d)
print(e)
print(f)
```
