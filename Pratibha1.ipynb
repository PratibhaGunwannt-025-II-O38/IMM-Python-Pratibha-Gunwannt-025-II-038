```python
#checking data types

print(type(10))   #int
print(type(3.14))  #float
print(type(1+3j))  #complex
print(type('Pratibha'))  #string
print(type[1, 2, 3])  #list
print(type({'name': 'Pratibha'}))  #dictionary
print(type({9.8, 3.144, 2.7}))  #tuple
# list is mutable(can be modified) while tuple isn't
```
