```python
radius = 10
pi = 3.14

# Use ** for power (exponent)
area = pi * (radius ** 2) 

# Use {} as placeholders for the .format() method
result = 'The area of circle with {} is {}'.format(radius, area)

print(result) 
# Output: The area of circle with 10 is 314.0
print(area)
```
